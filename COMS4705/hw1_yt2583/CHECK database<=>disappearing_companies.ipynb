```python
##################################################################################
#          Copyright (c) 2018 by IoT Nation LLC. All Rights Reserved.            #
##################################################################################
#           IoT Nation Confidential & Proprietary Information                    #
##################################################################################
# NOTICE: All information contained herein is, and remains the intellectual      #
# property of IoT Nation LLC. The intellectual, technical, and business concepts #
# contained herein including but not limited to algorithms, source code,         #
# executables and derivative files or other information resulting from executing #
# this code are proprietary to IoT Nation LLC and may be covered by U.S. and     #
# foreign patents and/or patents in process, and are protected by trade secret   #
# or copyright law.                                                              #
#                                                                                #
# The copyright notice above does not evidence any actual or intended            #
# publication or disclosure of this source code. Dissemination, reproduction     #
# (whether digitally or physically), or use of any part of this material is      #
# strictly forbidden unless prior written permission is obtained from the CEO of #
# IoT Nation LLC. The receipt or posession of this source code and/or related    #
# information does not convey or imply any rights to reproduce, disclose or      #
# distribute its contents, or to manufacture, use, or sell anything that it may  #
# describe, in whole or in part.                                                 #
##################################################################################


import psycopg2
from pprint import pprint
from common.database.db_conn import *
from common.database.config import config  
import csv
import json
import pandas as pd
import sys
import os
from sys import path
path = '../csvfiles/'
path2 = '../inifiles/'
class DBConnection:
    def __init__(self,infile):
        try:
            params = config(infile)
            self.connection = psycopg2.connect(**params)
            self.connection.autocommit = True
            self.cursor = self.connection.cursor()
        except:
            pprint("Cannot connect to database")
            
    def execute_command(self,command):
        try:
            self.cursor.execute(command)
        except:
            print 'Error！！！'+command
    
    def find_attributes(self,table_name):
        self.cursor.execute("Select * FROM {0}".format(table_name))
        colnames = [desc[0] for desc in self.cursor.description]
        return colnames
    
    def insert_new_record(self,table_name,new_record,fill_attributes):#row by row insertion
        #new_record = table_attributes
        i = 0
        try:
            
            record = "INSERT INTO {0}".format(table_name) + fill_attributes + " VALUES('" 
            for j in range(0,len(new_record)-1):
                bug = new_record[j].replace('\'','\'\'').strip()
                record = record + bug + "','"
            bug2 = new_record[-1].replace('\'','\'\'').strip()
            record = record + bug2 + "')"
            insert_command = record.replace("'null'","null")
            #pprint(insert_command)
            #print insert_command
            i+=1
            try:
                self.cursor.execute(insert_command)
            except:
                print 'Error！！！'+insert_command
        except:
            print i
        
    def insert_new_record_lowercase(self,table_name,new_record,fill_attributes):#row by row insertion
        #new_record = table_attributes
        record = "INSERT INTO {0}".format(table_name) + fill_attributes + " VALUES('" 
        for j in range(0,len(new_record)-1):
            bug = new_record[j].replace('\'','\'\'').strip()
            record = record + bug + "','"
        bug2 = new_record[-1].replace('\'','\'\'').strip()
        record = record + bug2 + "')"
        insert_command = record.replace("'null'","null").lower()
        #pprint(insert_command)
        #print insert_command
        try:
            self.cursor.execute(insert_command)
        except:
            print 'Error！！！'+insert_command

    def query_all(self,table_name):
        self.cursor.execute("SELECT * FROM {0}".format(table_name))
        tuple_all = self.cursor.fetchall()
        return tuple_all
        #for tuple_each in tuple_all:
            #pprint("each_tuple : {0}".format(tuple_each))

    def drop_table(self,table_name):
        drop_table_command = "DROP TABLE {0}".format(table_name)
        self.cursor.execute(drop_table_command)


def gen_realtion_attribute(fieldnames):
    fill_attributes = '('
    for i in range(0,len(fieldnames)-1):
        fill_attributes += fieldnames[i]+','
    fill_attributes = fill_attributes + fieldnames[-1] + ')'
    return fill_attributes


def fetch_database_data(infile, command, *values):
    params = config(infile)
    conn = psycopg2.connect(**params)
    cur = conn.cursor()
    if not values:        
        try:
            cur.execute(command)      
            rows = cur.fetchall()
            return rows
        except psycopg2.Error as e:
            print e
    else:
        try:
            cur.execute(command, values)      
            rows = cur.fetchall()
            return rows
        except psycopg2.Error as e:
            print e        
    cur.close()
    conn.close()

from_knack = pd.DataFrame.from_csv(path+'iotnationbd.csv')
from_knack
```

```python
database_from = path2+'database.ini'
com_name = fetch_database_data(database_from, "select cu.company_id, cu.company_name, s.source_confidence from company_name_table cu, source_table s where s.source_id = cu.source_id;")
com_name = pd.DataFrame.from_records(com_name,columns=['Company ID','Company Name', 'con']).drop_duplicates()
com_name = com_name.loc[com_name.groupby('Company ID').idxmax('con')['con']][['Company ID','Company Name']]
com_url = fetch_database_data(database_from, "select cu.company_id, cu.company_url, s.source_confidence from company_url_table cu, source_table s where s.source_id = cu.source_id;")
com_url = pd.DataFrame.from_records(com_url,columns=['Company ID','Company URL', 'con']).drop_duplicates()
com_url = com_url.loc[com_url.groupby('Company ID').idxmax('con')['con']][['Company ID','Company URL']]
com_iot_rank = fetch_database_data(database_from, "select cu.company_id, cu.company_iot_rank, s.source_confidence from company_iot_nation_rank cu, source_table s where s.source_id = cu.source_id;")
com_iot_rank = pd.DataFrame.from_records(com_iot_rank,columns=['Company ID','IoTN Rank', 'con']).drop_duplicates()
com_iot_rank = com_iot_rank.loc[com_iot_rank.groupby('Company ID').idxmax('con')['con']][['Company ID','IoTN Rank']]
com_size = fetch_database_data(database_from, "select cu.company_id, cu.company_size, s.source_confidence from company_size_table cu, source_table s where s.source_id = cu.source_id;")
com_size = pd.DataFrame.from_records(com_size,columns=['Company ID','Company Size', 'con']).drop_duplicates()
com_size = com_size.loc[com_size.groupby('Company ID').idxmax('con')['con']][['Company ID','Company Size']]
com_phone = fetch_database_data(database_from, "select cu.company_id, cu.company_phone, s.source_confidence from company_phone_table cu, source_table s where s.source_id = cu.source_id;")
com_phone = pd.DataFrame.from_records(com_phone,columns=['Company ID','Company Phone', 'con']).drop_duplicates()
com_phone = com_phone.loc[com_phone.groupby('Company ID').idxmax('con')['con']][['Company ID','Company Phone']]
com_phone['Company Phone'] = com_phone['Company Phone'].str.replace('+', '')
com_email = fetch_database_data(database_from, "select cu.company_id, cu.company_email, s.source_confidence from company_email_table cu, source_table s where s.source_id = cu.source_id;")
com_email = pd.DataFrame.from_records(com_email,columns=['Company ID','Company Email', 'con']).drop_duplicates()
com_email = com_email.loc[com_email.groupby('Company ID').idxmax('con')['con']][['Company ID','Company Email']]
com_address = fetch_database_data(database_from, "select a.company_id, c.street_1, c.street_2, c.city, c.state, c.zipcode, c.country, s.source_confidence from address_table c, company_address_table a, source_table s where s.source_id = a.source_id and a.address_id = c.address_id and a.company_address_type = 'hq';")
com_address = pd.DataFrame.from_records(com_address,columns=['Company ID','s1', 's2', 'Company City', 'Company State', 'zipcode', 'Company Country', 'con']).drop_duplicates()
com_address = com_address.loc[com_address.groupby('Company ID').idxmax('con')['con']][['Company ID', 's1', 's2', 'Company City', 'Company State', 'zipcode', 'Company Country']]
com_address = com_address.fillna('')
com_address['Company Address'] = com_address['s1'] + ' ' + com_address['s2'] + ' ' + com_address['Company City'] + ' ' +com_address['Company State'] + ' ' + com_address['zipcode'] + ' ' + com_address['Company Country']
com_address = com_address[['Company ID', 'Company Address', 'Company City', 'Company State','Company Country']]
com_description = fetch_database_data(database_from, "select cu.company_id, cu.company_description, s.source_confidence from company_description_table cu, source_table s where s.source_id = cu.source_id;")
com_description = pd.DataFrame.from_records(com_description,columns=['Company ID','Company Description', 'con']).drop_duplicates()
com_description = com_description.loc[com_description.groupby('Company ID').idxmax('con')['con']][['Company ID','Company Description']]
com_linkedin = fetch_database_data(database_from, "select cu.company_id, cu.company_linkedin, s.source_confidence from company_linkedin_table cu, source_table s where s.source_id = cu.source_id;")
com_linkedin = pd.DataFrame.from_records(com_linkedin,columns=['Company ID','Company Linkedin', 'con']).drop_duplicates()
com_linkedin = com_linkedin.loc[com_linkedin.groupby('Company ID').idxmax('con')['con']][['Company ID','Company Linkedin']]
com_twitter = fetch_database_data(database_from, "select cu.company_id, cu.company_twitter, s.source_confidence from company_twitter_table cu, source_table s where s.source_id = cu.source_id;")
com_twitter = pd.DataFrame.from_records(com_twitter,columns=['Company ID','Company twitter', 'con']).drop_duplicates()
com_twitter = com_twitter.loc[com_twitter.groupby('Company ID').idxmax('con')['con']][['Company ID','Company twitter']]
com_facebook = fetch_database_data(database_from, "select cu.company_id, cu.company_facebook, s.source_confidence from company_facebook_table cu, source_table s where s.source_id = cu.source_id;")
com_facebook = pd.DataFrame.from_records(com_facebook,columns=['Company ID','Company Facebook', 'con']).drop_duplicates()
com_facebook = com_facebook.loc[com_facebook.groupby('Company ID').idxmax('con')['con']][['Company ID','Company Facebook']]
com_news = fetch_database_data(database_from, "select cu.company_id, cu.company_news, s.source_confidence from company_news_table cu, source_table s where s.source_id = cu.source_id;")
com_news = pd.DataFrame.from_records(com_news,columns=['Company ID','Company News', 'con']).drop_duplicates()
com_news = com_news.loc[com_news.groupby('Company ID').idxmax('con')['con']][['Company ID','Company News']]
com_blog = fetch_database_data(database_from, "select cu.company_id, cu.company_blog, s.source_confidence from company_blog_table cu, source_table s where s.source_id = cu.source_id;")
com_blog = pd.DataFrame.from_records(com_blog,columns=['Company ID','Company Blog', 'con']).drop_duplicates()
com_blog = com_blog.loc[com_blog.groupby('Company ID').idxmax('con')['con']][['Company ID','Company Blog']]
df = pd.merge(com_name, com_iot_rank, on ='Company ID', how = 'left')
df = pd.merge(df, com_url, on ='Company ID', how = 'left')
df = pd.merge(df, com_phone, on ='Company ID', how = 'left')
df = pd.merge(df, com_email, on ='Company ID', how = 'left')
df = pd.merge(df, com_size, on ='Company ID', how = 'left')
df = pd.merge(df, com_address, on ='Company ID', how = 'left')
df = pd.merge(df, com_description, on ='Company ID', how = 'left')
df = pd.merge(df, com_linkedin, on ='Company ID', how = 'left')
df = pd.merge(df, com_twitter, on ='Company ID', how = 'left')
df = pd.merge(df, com_facebook, on ='Company ID', how = 'left')
df = pd.merge(df, com_news, on ='Company ID', how = 'left')
df = pd.merge(df, com_blog, on ='Company ID', how = 'left')
df1 = df.sort_values('Company ID').reset_index(drop=True)
df1

```
